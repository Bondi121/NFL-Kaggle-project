## Analyzing tackle range including angles of pursuit, speed, acceleration, and closing speed.

- Identify Relevant Plays and Players: From the final_merged_data, we'll identify plays where tackles occur, along with the players involved (tacklers and ball carriers).

- Extract Tracking Data for Those Plays: Using gameId and playId, we can link these plays to the all_tracking_data to get the detailed positional and movement data of the players involved.

- Calculate Angle of Pursuit: We'll need to determine the angle at which the tackler approaches the ball carrier. This involves calculating the direction of the tackler relative to the ball carrier's position.

- Analyze Speed and Acceleration: We can use the speed (s) and acceleration (a) data from all_tracking_data to analyze how the speed and acceleration of the tackler changes during the pursuit.

- Determine Closing Speed: Closing speed can be assessed by how quickly the distance between the tackler and the ball carrier decreases, especially in the moments leading up to the tackle.

```python
import pandas as pd

# Assuming you've already loaded your datasets
# final_merged_data = pd.read_csv('path_to_final_merged_data.csv')
# all_tracking_data = pd.read_csv('path_to_all_tracking_data.csv')

# Filter plays with tackles
tackle_plays = final_merged_data[final_merged_data['tackle'] == 1]

# Select relevant columns
tackle_plays = tackle_plays[['gameId', 'playId_x', 'nflId', 'displayName', 'ballCarrierId', 'ballCarrierDisplayName']]

```

```python
# Extract ball carrier names
ball_carrier_names = final_merged_data[['nflId', 'displayName']].drop_duplicates()

# Merge to get ball carrier display names
tackle_plays = pd.merge(tackle_plays, ball_carrier_names, left_on='ballCarrierId', right_on='nflId', how='left')

# Rename columns for clarity
tackle_plays.rename(columns={'displayName_x': 'tacklerName', 'displayName_y': 'ballCarrierDisplayName'}, inplace=True)

# Drop the extra nflId column
tackle_plays.drop(columns=['nflId_y'], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is already loaded
# final_merged_data = pd.read_csv('path_to_final_merged_data.csv')

# Filter plays with tackles
tackle_plays = final_merged_data[final_merged_data['tackle'] == 1]

# Select relevant columns
tackle_plays = tackle_plays[['gameId', 'playId_x', 'nflId', 'displayName', 'ballCarrierId', 'ballCarrierDisplayName']]

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Visualize top tacklers
top_tacklers = tackle_plays['displayName'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_tacklers.values, y=top_tacklers.index, palette='coolwarm')
plt.title('Top 10 Tacklers')
plt.xlabel('Number of Tackles')
plt.ylabel('Player Name')
plt.show()

# Visualize top players who were tackled
top_tackled = tackle_plays['ballCarrierDisplayName'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_tackled.values, y=top_tackled.index, palette='coolwarm')
plt.title('Top 10 Players Tackled')
plt.xlabel('Number of Times Tackled')
plt.ylabel('Player Name')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter plays with tackles
tackle_plays = final_merged_data[final_merged_data['tackle'] == 1]

# Tackles by Quarter
tackles_by_quarter = tackle_plays['quarter'].value_counts()
sns.barplot(x=tackles_by_quarter.index, y=tackles_by_quarter.values, palette="viridis")
plt.title('Tackles by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Number of Tackles')
plt.show()

# Tackles by Down
tackles_by_down = tackle_plays['down'].value_counts()
sns.barplot(x=tackles_by_down.index, y=tackles_by_down.values, palette="magma")
plt.title('Tackles by Down')
plt.xlabel('Down')
plt.ylabel('Number of Tackles')
plt.show()
```

```python
# Tackles by Defensive Team
tackles_by_def_team = tackle_plays['defensiveTeam'].value_counts()
sns.barplot(x=tackles_by_def_team.values, y=tackles_by_def_team.index, palette="coolwarm")
plt.title('Tackles by Defensive Team')
plt.xlabel('Number of Tackles')
plt.ylabel('Defensive Team')
plt.show()
```

```python
# Analyzing relationship between tackles and forced fumbles
forced_fumbles = tackle_plays[tackle_plays['forcedFumble'] == 1]
sns.countplot(x='down', data=forced_fumbles, palette='Set2')
plt.title('Forced Fumbles by Down')
plt.xlabel('Down')
plt.ylabel('Number of Forced Fumbles')
plt.show()
```
